Lecture 8 — Gradient Descent Algorithm

Importing Libraries

```python
import pandas as pd 
import numpy as np  
import matplotlib.pyplot as plt 
```

Loading Values

```python
data = pd.read_csv("datasets/lecture6_salaryData.csv")
```

Extracting Values

```python
X = data["Experience Years"].values
y = data["Salary"].values
```

Converting X and y Into Numpy Arrays

```python
X = np.array(X)
y = np.array(y)
X
```

Plot dataset

```python
plt.scatter(X,y)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Experience VS Salary")
plt.show()
```

Initialiing Parameters For GD

```python
m = 0
b = 0

learning_rate = 0.01
iterations = 1000

n = len(X)
```

Computing Cost Function

```python
def cost_func(X,y,m,b):
    
    y_pred = m * X + b
    cost = np.mean((y-y_pred)**2)
    
    return cost
```

Implementing Gradient Descent

```python
cost_history = []

for i in range(iterations):
    y_pred = m*X+b
    
    dm = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)
    
    m = m - learning_rate * dm
    b = b - learning_rate * db
    
    cost = cost_func(X,y,m,b)
    cost_history.append(cost)
```

Final Parameters

```python
print("Slope(m) : ",m)
print("Intercept(b) : ",b)
```

Prediction

```python
y_pred = m * X + b
```

Plotting Regression Line

```python
plt.scatter(X,y)

plt.plot(X,y_pred, color = "red")
plt.xlabel("Experience")
plt.ylabel("Salary")

plt.title("Linear Regression using Gradient Descent")

plt.show()
```

Cost Function Convergence

```python
plt.plot(range(iterations), cost_history)

plt.xlabel("Iterations")
plt.ylabel("Cost")

plt.title("Cost Function Convergence")

plt.show()
```

Test Prediction

```python
experience = int(input("Enter The Experience : "))
pred = m * experience + b
print("Expected Salary : ",pred)
```
