**1. Regression Problems**

Regression is used when the output is a number.

Example outputs:

House price → ₹45,00,000

Temperature → 32°C

Salary → ₹8,50,000

Stock price → ₹320

So if the target variable is numeric, it is a Regression problem.

**2. Classification Problems**

Classification is used when the output is a category or label.

Examples:

Spam or Not Spam

Disease or Healthy

Fraud or Not Fraud

Cat or Dog

Output is discrete categories.

**3. Clustering Problems**

Clustering is used when data has no labels.

The algorithm automatically groups similar data together.

This is Unsupervised Learning.

**4. Dimensionality Reduction**

Sometimes datasets contain too many features.

Example dataset:

| Feature1 | Feature2 | Feature3 | Feature4 | Feature5 |

Many features can cause:

slow models

overfitting

high complexity

Dimensionality reduction reduces the number of features.
