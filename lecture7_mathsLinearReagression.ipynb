Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Loading Data

```python
data = pd.read_csv("datasets/lecture6_salaryData.csv")
data.head()
```

Dividing feature value and target value

```python
X = data["Experience Years"]
y = data["Salary"]
```

Plot Dataset

```python
plt.scatter(X,y)

plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Experience Vs Salary")

plt.show()
```
